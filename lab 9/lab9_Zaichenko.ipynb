**Наївний пошук**

```python
def naive_match(p, t):
    assert len(p) <= len(t) 
    occurrences = []
    for i in range(0, len(t)-len(p)+1): 
        match = True
        for j in range(0, len(p)): 
            if t[i+j] != p[j]:
                match = False 
                break
        if match:
            occurrences.append(i)
    return occurrences

#перевірка 
text = "abababab"
pattern = "ab"
result = naive_match(pattern, text)

print("\nТекст:", text)
print("Шаблон:", pattern)
print("Позиції входження:", result)
```

Складність у найгіршому випадку: 𝑂(|𝑡| ⋅ |𝑝|)

**Z-функція**


```python
def z_func(s):
    n = len(s)
    z = [0] * n
    l = r = 0

    for i in range(1, n):
        if i <= r:
            z[i] = min(r - i + 1, z[i - l])

        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1

        if i + z[i] - 1 > r:
            l = i
            r = i + z[i] - 1
    z[0] = n
    return z

#перевірка 
string = "abababab"
result = z_func(string)
print(result)
```

**Пошук через Z-функцію**

```python
def z_match(p, t):
    s = p + "$" + t
    z = z_func(s)
    res = []

    for i in range(len(p) + 1, len(s)):
        if z[i] == len(p):
            res.append(i - len(p) - 1)

    return res

#перевірка 
t, p = "абабагаламага", "аб"
result = z_match(p, t)
print(result)
```

Асимптотична складність алгоритму: 𝑂(𝑙𝑒𝑛(𝑝) + 𝑙𝑒𝑛(𝑡)).


**Стиснення рядка**

```python
def compress_with_z(s):
    z = z_func(s)
    n = len(s)

    for i in range(1, n):
        if i + z[i] == n and n % i == 0:
            return s[:i]
    return s

#перевірка
string = "abababab"
compressed = compress_with_z(string)
print(compressed)
```

**1. Що таке «префікс-функція» у контексті алгоритмів на рядках, і як вона відрізняється від Z-функції?**

Префікс-функція - це масив, де для кожної позиції показується довжина найбільшого власного префікса рядка, який одночасно є суфіксом для підрядка, що закінчується в цій позиції.

* π-функція → дивиться “назад” (префікс = суфікс)
* Z-функція → дивиться “вперед” (порівняння з початком)

**2. Що таке Z-функція у вищому контексті алгоритмів на рядках, і яка її роль у вирішенні задач?** 

Z-функція - це масив, який для кожної позиції показує довжину найбільшого підрядка, що співпадає з префіксом рядка.

Вона дозволяє:
* швидко знаходити підрядки
* аналізувати структуру рядка
* знаходити повтори
* стискати рядки

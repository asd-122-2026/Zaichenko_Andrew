```python
global heap
global currSize

heap = [0]      
currSize = 0

def parent(i): 
 return i // 2
def left(i): 
 return 2*i
def right(i): 
 return (2*i + 1)
# міняемо елемент з індексом a на елемент з індексом b
def swap(a, b):
    temp = heap[a]
    heap[a] = heap[b]
    heap[b] = temp
# додавання елемента
def insert(elem):

    global currSize
    index = len(heap)
    heap.append(elem)
    currSize += 1
    par = parent(index)
    flag = 0

    while flag != 1:
        if index == 1:
            flag = 1
        elif heap[par] > elem:
            flag = 1
        else:
            swap(par, index)
            index = par
            par = parent(index)
def maxHeapify(index):

    lar = index
    l = left(index)
    r = right(index)
    if l <= currSize and heap[l] > heap[lar]:
        lar = l
    if r <= currSize and heap[r] > heap[lar]:
        lar = r
    if lar != index:
        swap(index, lar)
        maxHeapify(lar)
# видалення макс.елемента
def extractMax():

    global currSize 
    if currSize != 0:
        maxElem = heap[1]
        heap[1] = heap[currSize]
        heap.pop(currSize)
        currSize -= 1
        maxHeapify(1)
        return maxElem
# генерація купи
import random

arr = []
for i in range(10):
    arr.append(random.randint(1, 50))
print(arr)
# заповнюємо
for x in arr:
    insert(x)
print(heap)
# додаю номер варіанту
insert(9)
print(heap)
#викликаю функцію
maxElem = extractMax()
print(maxElem)
print(heap)

```

#### Завдання 3

– Реалізуйте геш-таблицю з ланцюжковим гешуванням.
– Проведіть тестування геш-таблиці з різними типами даних (цілі числа, рядки, списки, словники, об’єкти):
* перевірити працездатність геш-таблиці з різними типами даних.
* виміряти час виконання основних операцій (пошук, вставка, видалення) для різних типів даних.
* порівняти результати для різних типів даних.

```python
class HashTable:

    def __init__(self, size):
        self.size = size
        self.table = []
        for i in range(size):
            self.table.append([])

    def hash_function(self, key):
        return hash(str(key)) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        bucket = self.table[index]
        for k, v in bucket:
            if k == key:
                return v
        return None

    def delete(self, key):

        index = self.hash_function(key)
        bucket = self.table[index]
        for i in range(len(bucket)):
            if bucket[i][0] == key:
                del bucket[i]
                return True
        return False
# перевірка
ht = HashTable(10)

ht.insert(10, "text")
print(ht.search(10))

ht.insert("key", 123)
ht.insert("11", [1, 2, 3])
ht.insert((1, 2), {"a": 1})

print(ht.search("key"))
print(ht.search("11"))
print(ht.search((1, 2)))
```

3) Повторювати, поки властивість купи не відновиться.

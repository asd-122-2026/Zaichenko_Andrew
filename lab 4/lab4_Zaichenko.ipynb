```python
#Обчислення факторіала за допомогою цикла for
def factorial_for(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial_for(5))
```

Цикл виконується n разів, тому часова складність: O(n)

```python
# Обчислення факторіала через рекурсію
def fac(n):
    if n == 0:
        return 1
    return fac(n - 1) * n
print(fac(5))
```

Рекурсивна функція викликає себе n разів, для кожного виклику виконується одна операція множення.Тому часова складність алгоритму: O(n)

```python
#Рекурсивне обчислення n-го числа Фібоначчі
def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(30))
```

Часова складність такого алгоритму: O(2^n), тобто алгоритм має експоненційну складність.

```python
#Зливає два відсортованих масиви left і right у один
def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(arr):
    if len(arr) < 2:
        return arr
    middle = len(arr) // 2
    left = mergesort(arr[:middle])
    right = mergesort(arr[middle:])
    return merge(left, right)
a = [6, 5, 3, 1, 8, 7, 2, 4]
merge([2, 4, 5, 7], [1, 2, 3, 6]) 
[1, 2, 2, 3, 4, 5, 6, 7] 
mergesort(a)
```

Кількість рівнів поділу масиву дорівнює logn, оскільки масив щоразу ділиться навпіл. Отже, загальна часова складність алгоритму дорівнює: O(nlogn)

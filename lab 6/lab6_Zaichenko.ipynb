```python
# У стилі ООП
class Stack:
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def peek(self):
        return self.items[-1]
    def size(self):
        return len(self.items)
s = Stack()
s.push('hello')
s.push('true')
print(s.pop())  
print(s.pop())  
```

**написати функцію pop_n(), що видаляє елементи стека з його початку до номера n включно;**

```python
def pop_n(stack, n):
    for i in range(n + 1):
        if len(stack) > 0:
            stack.pop()
stack = [10, 20, 30, 40, 50]
pop_n(stack, 2)
print(stack)
```

### **Контрольні питання**

**1. Що таке стек і які операції можна виконувати зі стеком?**



Стек — це структура даних, яка працює за принципом LIFO (Last In — First Out, «останній прийшов — перший вийшов»).


Основні операції зі стеком:
* push — додавання елемента у вершину стека;
* pop — видалення елемента з вершини;
* peek або top — перегляд верхнього елемента без видалення;
* isEmpty — перевірка, чи порожній стек;
* size — визначення кількості елементів.

**2. Яка основна відмінність між стеком та чергою?**

| Стек                                              | Черга                                       |
| ------------------------------------------------- | ------------------------------------------- |
| Працює за принципом **LIFO**                      | Працює за принципом **FIFO**                |
| Останній доданий елемент видаляється першим       | Перший доданий елемент видаляється першим   |
| Додавання і видалення відбувається з одного кінця | Додавання — в кінець, видалення — з початку |


**3. Як ви можете реалізувати стек за допомогою масиву і за допомогою зв’язаного списку? Які переваги та недоліки кожного підходу?**

Реалізація стеку за допомогою масиву. Елементи зберігаються у звичайному масиві або списку.

Переваги:
* проста реалізація;
* швидкий доступ до елементів;
* операції push і pop виконуються швидко.


Недоліки:
* фіксований розмір масиву;
* можливе переповнення;
* інколи потрібно перевиділення пам’яті.

Реалізація стеку за допомогою зв’язаного списку. Кожен елемент містить значення та посилання на наступний вузол.

Переваги:
* динамічний розмір;
* ефективне використання пам’яті;
* немає переповнення фіксованого масиву.

Недоліки:
* складніша реалізація;
* додаткові витрати пам’яті на посилання.

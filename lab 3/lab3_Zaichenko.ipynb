##### **Завдання 1: Виконати теоретичні розрахунки асимптотичної складності алгоритмів сорутвання методами вставки та бульбашки.**

Сортування методом вставками:

* Найкраща асимптотична складність - O(n) лінійна, оскільки масив уже відсортований і потрібні лише порівняння.
* Найгірша асимптотична складність - O(n^2) квадратична, тому що елементи доводиться багаторазово зсувати.

Сортування методом бульбашки:

* Найкраща асимптотична складність - O(n) лінійна, якщо є перевірка обмінів і масив уже відсортований.
* Найгірша асимптотична складність - O(n^2) квадратична, оскільки виконується багато проходів і порівнянь сусідніх елементів.

##### **Завдання 2: Реалізувати на Python алгоритм сортування методом бульбашки.**

```python
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):

        swapped = False
        for j in range(0, n - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

    return arr

numbers = [7, 3, 10, 2, 8, 1]
start_time = time.time()
sorted_numbers = bubble_sort(numbers)
print("Відсортований масив:")
print(sorted_numbers)
print("\nЧас виконання:")
print("--- %s seconds ---" % (time.time() - start_time))
```

##### **Завдання 3: Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей 𝑇𝑖𝑛𝑠𝑒𝑟𝑡(𝑛) та 𝑇𝑏𝑢𝑏𝑙𝑒(𝑛) для 𝑛 = [5, 10,50,100,500,1000,2000,3000,4000,5000,10000,20000,50000,10000], згенерувавши відповідні цілочисельні масиви.**

```python
import random
import time
import matplotlib.pyplot as plt

def insertion_sort(arr):
    a = arr.copy()

    for i in range(1, len(a)):
        key = a[i]
        j = i - 1

        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1

        a[j + 1] = key

    return a

def bubble_sort(arr):
    a = arr.copy()
    n = len(a)

    for i in range(n):
        swapped = False

        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
                swapped = True

        if not swapped:
            break

    return a

sizes = [5, 10, 50, 100, 500, 1000, 2000]
insert_times = []
bubble_times = []

for n in sizes:

    arr = [random.randint(0, 10000) for _ in range(n)]

    start = time.time()
    insertion_sort(arr)
    end = time.time()

    insert_times.append(end - start)

    start = time.time()
    bubble_sort(arr)
    end = time.time()

    bubble_times.append(end - start)

plt.plot(sizes, insert_times, marker='o', label='Insertion Sort')
plt.plot(sizes, bubble_times, marker='o', label='Bubble Sort')

plt.xlabel("Розмір масиву n")
plt.ylabel("Час виконання (сек)")
plt.title("Порівняння алгоритмів сортування")

plt.legend()
plt.grid()

plt.show()
```

#### **Контрольні запитання**

**1. Що таке O-нотація і чим вона відрізняється від Ω-нотації?**
* $O$-нотація описує верхню межу часу виконання. Це «найгірший сценарій» — гарантія того, що алгоритм не працюватиме повільніше за певний рівень. 
* $\Omega$-нотація описує нижню межу. Це «найкращий сценарій» — швидкість, нижче якої алгоритм точно не опуститься.
**2. Яку часову складність має алгоритм сортування методом вставляння за найсприятливіших умов?**
Коли масив уже відсортований, складність алгоритму становить $O(N)$.
**3. Визначення ефективного алгоритму**
Ефективний алгоритм - це алгоритм, який виконує задачу з мінімальними витратами часу та пам’яті, тобто має оптимальну асимптотичну складність для даної задачі.
**4. Пояснити поняття головного параметра (розміру) задачі.**
Це кількісна характеристика вхідних даних, від якої залежить час виконання алгоритму. Зазвичай позначається як $N$. 
* У задачах сортування — це кількість елементів у масиві.
* У задачах з графами — це кількість вершин або ребер.
* У роботі з текстом — це кількість символів.
